## A notebook for analyzing flight records from a day in April, 2024. 

```python
# Python imports
import pandas as pd
import matplotlib.pyplot as plt
# import numpy as np
# import ipywidgets as widgets
```

```python
# Import data, read from JSON file
df = pd.read_json('./one-day-flights.json')

# Note column names
print(df.columns)
```

```python
# Sort data, based on "estArrivalAirport" column
sorted_by_dest = df.sort_values(by=['estArrivalAirport'])

# Sort data, based on "estArrivalAirport" column
sorted_by_depart= df.sort_values(by=['estDepartureAirport'])

# display the sorted data
print(sorted_by_dest)

print(sorted_by_depart)
```

```python
# Create a chart showing frequency of destination airports
# Count frequency of each airport destination
frequency = df['estArrivalAirport'].value_counts()

# Plot the frequency as a bar chart
frequency.plot(kind='bar')

# Show the plot
plt.show()
```

```python
# Print top 20 destinations and their frequency (in this day of data)

# Print the top 20 rows
print(frequency.head(20))
```

```python
# Save top 20 rows in new df
top_20 = frequency.head(20)

# Plot pie chart by location
# Define the data
data = {'Airport': ['RJTT', 'LEMD', 'EDDM', 'VOBL', 'VTBS', 'EHAM', 'OMDB', 'LTFJ', 'VIDP', 'YSSY', 'ENGM', 'LEBL', 'YBBN', 'ESSA', 'YPJT', 'LTFM', 'LFPG', 'FAOR', 'LEPA', 'RPLL'],
        'Frequency': [32, 26, 25, 22, 21, 20, 20, 19, 19, 18, 18, 18, 17, 17, 16, 15, 15, 14, 14, 14]}

# Convert the data to a pandas DataFrame
df = pd.DataFrame(data)

# Set the figure size
plt.figure(figsize=(10, 8))

# Create a pie chart
plt.pie(df['Frequency'], labels=df['Airport'], autopct='%1.1f%%')
plt.axis('equal')
plt.title('Airport Frequency')
plt.show()
```
